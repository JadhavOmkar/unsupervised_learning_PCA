## 2. Loading helper functions

```python
def generate_model_report(y_actual, y_predicted):
    """
    Calculate model evaluation parameters with input as actual Y column and predicted Y column
    """
    from sklearn.metrics import accuracy_score ## accuracy calculations
    from sklearn.metrics import precision_score, recall_score, confusion_matrix ## model evaluation parameters
    from sklearn.metrics import f1_score, roc_auc_score, roc_curve ##some more model evaluation parameters
    print("Accuracy = " , accuracy_score(y_actual, y_predicted))
    print("Precision = " ,precision_score(y_actual, y_predicted))
    print("Recall = " ,recall_score(y_actual, y_predicted))
    print("F1 Score = " ,f1_score(y_actual, y_predicted))
    pass 
```

```python
def quick_report(df):

  """ 
  This function will give a quick overview about dataset shape, dtypes, column names and NULL value count
  
  """
  print('Data Types:')
  print(df.dtypes)
  print("----------------------------------------------------------------------")  
  print('Rows and Columns:')
  print(df.shape)
  print("----------------------------------------------------------------------")  
  print('Column Names:')
  print(df.columns)
  print("----------------------------------------------------------------------")  
  print('Number of Null Values:',df.isnull().sum().sum())
  #print(df.isnull().sum().sum())
  if df.isnull().sum().sum()>0:
    print("----------------------------------------------------------------------")  
    print('Null Values across columns:')
    print(df.apply(lambda x: sum(x.isnull())))  
```

```python
def corrank(X):
    """
    This function calculates Correlated pairs within the dataframe 
    """
    import itertools
    df = pd.DataFrame([[(i,j),X.corr().loc[i,j]] for i,j in list(itertools.combinations(X.corr(), 2))],columns=['pairs','corr'])    
    finaldf=df.sort_values(by='corr',ascending=False)
    return finaldf
```

```python
def scale_data(X):
    """
    This function will scale the data using StandardScaler function from Sklearn
    """
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    Y = scaler.fit_transform(X)
    Y=pd.DataFrame(Y,columns=[X.columns])
    return Y
```

```python
def vif_cal(input_data):
    
    """
    This funtion calculates VIF value from each variables within the dataframe
    """
    #x_vars=input_data.drop([dependent_col], axis=1)
    x_vars=input_data
    xvar_names=x_vars.columns
    for i in range(0,xvar_names.shape[0]):
        y=x_vars[xvar_names[i]]
        x=x_vars[xvar_names.drop(xvar_names[i])]
        rsq=ols(endog=y, exog=x, data=x_vars).fit().rsquared
        vif=round(1/(1-rsq),2)
        print (xvar_names[i], " VIF = " , vif)        
```

```python
def f_importances(coef, names):
    
    """
    This function will plot feature names and their feature importance score
    """
    imp = coef
    imp,names = zip(*sorted(zip(imp,names)))
    plt.barh(range(len(names)), imp, align='center')
    plt.yticks(range(len(names)), names)
    plt.show()
```

**Creating a dataframe in a manner that each column is placed in order of its importance score**

```python
# Creating a subset of dataframe with top 5 important variables according to the feature importance plot
X_fs=X[['elongatedness','max.length_aspect_ratio','scaled_radius_of_gyration.1'
        ,'max.length_rectangularity','pr.axis_aspect_ratio','compactness','skewness_about.2','scaled_radius_of_gyration'
        ,'skewness_about.1','skewness_about']]
```

**Creating a WHILE LOOP which will first create a 'X' dataframe initial with only one variable .i.e elongatedness and then after creating X_train and Y_train it will perform an grid search to find optimal parameters for the SVM model and apply the best score model on the X-train and Y-train will print score on train set and test set. Initial the loop will run for only one predictor which is 'elongatedness' then when n=2, the loop will run for two predictors which are 'elongatedness' & 'max.length_aspect_ratio' and so on till n=10.**
**Finally we will compare the result for each loop and consider the model with optimal number of predictors which will be efficient and cost effective**

```python
n = 1
while n < 11:
    subset=X_fs.iloc[:,0:n]
    X_train,X_test,Y_train,Y_test = train_test_split(subset, y, test_size=0.3, random_state=0)
    grid_search_SVM_fs = GridSearchCV(SVC(), param_grid, cv=3)
    grid_fit_SVM_fs=grid_search_SVM_fs.fit(X_train, Y_train)
    best_clf_SVM_fs=grid_fit_SVM_fs.best_estimator_
    grid_fit_SVM_fs.best_score_
    best_clf_SVM_fs.fit(X_train, Y_train)
    print('\033[1m Following score for dataframe with following predictor/predictors \033[0;0m',X_train.columns)
    print("\n")
    print("Training set score for SVM: %f" % best_clf_SVM_fs.score(X_train , Y_train))
    print("Testing  set score for SVM: %f" % best_clf_SVM_fs.score(X_test , Y_test ))
    print("\n")
    n=n+1
```

**From the output of the above while loop result, model score for dataframe with 5  predictors ('elongatedness', 'max.length_aspect_ratio', 'scaled_radius_of_gyration.1','max.length_rectangularity','pr.axis_aspect_ratio') is the optimum because the previous model score with initial 4 predictor had test set score of approx 93% and adding pr.axis_aspect_ratio' to the dataframe took the test score to 96% and when 6th predictor 'compactness' was added to the dataframe, the test set score reamined same. We can go ahead and create a final model with 5 predictors.**

```python
X_final=X[['elongatedness','max.length_aspect_ratio','scaled_radius_of_gyration.1'
        ,'max.length_rectangularity','pr.axis_aspect_ratio']]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_final, y, test_size=0.3, random_state=0)
```

```python
grid_search_SVM_opt = GridSearchCV(SVC(), param_grid, cv=3)
grid_fit_SVM_opt=grid_search_SVM_opt.fit(X_train, Y_train)
best_clf_SVM_opt=grid_fit_SVM_opt.best_estimator_
# best score
grid_search_SVM_opt.best_score_
```

```python
best_clf_SVM_opt.fit(X_train, Y_train)

# Make predictions using the new model.
best_train_predictions_opt = best_clf_SVM_opt.predict(X_train)
best_test_predictions_opt = best_clf_SVM_opt.predict(X_test)

print("Training set score for SVM: %f" % best_clf_SVM_opt.score(X_train , Y_train))
print("Testing  set score for SVM: %f" % best_clf_SVM_opt.score(X_test , Y_test ))
```

```python
print(confusion_matrix(Y_test,best_test_predictions_opt))
print("\n")
print(classification_report(Y_test,best_test_predictions_opt))
```

## 8. Applying Principal Component Analysis

```python
## Creating new dataframe which is scaled
X_pca=vif_data
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(X_pca)
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
List_of_explained_var=explained_variance.tolist()
```

```python
print("Percentage of variance explained by each principal component :")
for x in List_of_explained_var:
    print(round((float(x))*100,2),"%")
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(6, 4))

    plt.bar(range(18), explained_variance, alpha=0.5, align='center',
            label='individual explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
```

**Thus we can see from the above plot and result from the formloop that first two components constitute almost 71% of the variance.Third, fourth, fifth, sixth and seventh components has 25% of the data spread. That totals up to 95% of the total variance explained. The rest all component has less than 5% of the variance.Hence we can drop the rest components.** 
##### We will consider first 7 PC !!

```python
pca=PCA(n_components=7)
X_new=pca.fit_transform(X_pca)
#X_new
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

On the other hand, the third approach of using Principal components had successfully reduced the dimnesion of the data from 18 predictors to 7 PC expalinning 95% of the variability. But at the same time we have lost interpretability.
